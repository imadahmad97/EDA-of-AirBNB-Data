# A. Importing data and wrangling data
    1. read the 1st csv file named 'Listings.csv' as 'airbnb_raw'
    2. 'airbnb_raw' - change the data types for some of the columns
    3. 'airbnb_raw' - add column host_since_dt, host_since_dt_year
    3. 'airbnb_raw' - have a look on the data by grouping
    4. read the 2nd csv file name 'Reviews.csv' as 'airbnb_review_raw'
    5. 'airbnb_review_raw' - add column review_date_dt, day_from_last_review
    6. 'airbnb_review_raw' - add columns adding up the reviews by year_month from 2008_1 to 2021_12
    7. create dataframe 'airbnb_review_gb' grouping 'airbnb_review_raw' by 'listing_id'
    8. create dataframe 'airbnb_raw_plus_review' -  'airbnb_raw' left join 'airbnb_review_gb' by 'listing_id'
    9. create dataframe 'exchange_table_df' with city, currency, currency_rate
    10. 'airbnb_raw_plus_review' left join 'exchange_table_df'
    11. 'airbnb_review_raw'- add column price_USD

```python
# count the frequency of the items in amenities list (dict)
def CountFrequency(my_list):
 
    # Creating an empty dictionary
    freq = {}
    for item in my_list:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
 
    return freq
```
